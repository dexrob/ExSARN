```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# file_name_1 = '' # retraining
# file_name_2 = '' # direct training
def compare_plot(file_name_1, file_name_2):
    fig, ax = plt.subplots(figsize=(15, 7))

    data = pd.read_pickle('models_and_stats/'+file_name_1)
    epoch_train_loss = data['epoch_train_loss']
    ax.plot(epoch_train_loss[:], label='epoch_train_loss')
    epoch_val_loss = data['epoch_val_loss']
    ax.plot(epoch_val_loss[:], label='epoch_val_loss')
    epoch_train_acc = data['epoch_train_acc']
    ax.plot(epoch_train_acc[:], label='epoch_train_acc')
    epoch_val_acc = data['epoch_val_acc']
    ax.plot(epoch_val_acc[:], label='epoch_val_acc')
    test_acc1 = data['test_acc1']

    data = pd.read_pickle('models_and_stats/'+file_name_2)
    epoch_train_loss = data['epoch_train_loss']
    ax.plot(epoch_train_loss[:], label='_epoch_train_loss', linestyle='dashed')
    epoch_val_loss = data['epoch_val_loss']
    ax.plot(epoch_val_loss[:], label='_epoch_val_loss', linestyle='dashed')
    epoch_train_acc = data['epoch_train_acc']
    ax.plot(epoch_train_acc[:], label='_epoch_train_acc', linestyle='dashed')
    epoch_val_acc = data['epoch_val_acc']
    ax.plot(epoch_val_acc[:], label='_epoch_val_acc', linestyle='dashed')
    test_acc2 = data['test_acc1']

    ax.set_xlabel('epoch')
    ax.set_ylabel('acc&loss')
    ax.grid(True)
    ax.set_title('test acc after training'+str(test_acc1)+'/'+str(test_acc2))
    fig.legend(loc='upper right')

# plt.show()
# plt.savefig('save/'+model_name+'stats.png')
```

```python
# c20
file_name_1 = 'server_models/stiffness_roughness_B_block_LSTM_data_c20_wrI_0.01_wC_1_wp_0.0_hidden_90_latent_40_r_on_0.1_p_max_200_stats_fold0_0.pkl' # retraining
file_name_2 = 'server_models/stiffness_roughness_B_block_LSTM_data_c20_wrI_0.01_wC_1_wp_0.1_hidden_90_latent_40_r_on_0.1_p_max_200_stats_fold0_0.pkl' # direct training
compare_plot(file_name_1, file_name_2)
```

```python
# c20new
file_name_1 = 'server_models/stiffness_roughness_B_block_LSTM_data_c20new_wrI_0.01_wC_1_wp_0.0_hidden_90_latent_40_r_on_0.1_p_max_200_stats_fold0_0.pkl' # retraining
file_name_2 = 'server_models/stiffness_roughness_B_block_LSTM_data_c20new_wrI_0.01_wC_1_wp_0.5_hidden_90_latent_40_r_on_0.1_p_max_200_stats_fold0_0.pkl' # direct training
compare_plot(file_name_1, file_name_2)
```
