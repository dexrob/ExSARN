```python
#!/usr/bin/env python
# coding: utf-8

'''
sample command: python T4_BT19_ae.py -k 0 -c 0 -r 1 --data_dir /home/ruihan/data
Individual training for BioTac data (full/partial data)
if -r=1, train with full data
if -r=2, train with half data
loss = classification loss + recon loss 
'''

# Import
import os,sys
import pickle
import argparse
from datetime import datetime
import numpy as np
import torch
import torch.nn as nn
import torchvision
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt
import seaborn as sns

from vrae.vrae import VRAEC, VRAEC_v2
# c20_2: preprocess_data_copy
from preprocess_data import get_TrainValTestLoader, get_TrainValTestDataset, get_TrainValTestData, property_label, c30_process_data
from vrae.visual import plot_grad_flow, plot_stats, tsne, multiple_tsne, confusion_matrix_show

def train(args):
    # Set hyper params
    args_data_dir = args.data_dir
    kfold_number = args.kfold
    data_reduction_ratio = args.reduction
    shuffle = True # set to False for partial training
    sequence_length = args.sequence_length
    number_of_features = 19

    hidden_size = args.h_s
    hidden_layer_depth = 1
    latent_length = args.latent_length
    batch_size = 32
    learning_rate = 0.001 # 0.0005
    n_epochs = 100
    dropout_rate = 0.2
    cuda = False # options: True, False
    header = None
    dataset = args.dataset
    if dataset == 'c50':
        num_class_texture = 50
    else:
        num_class_texture = 20
    
    use_stiffness = args.use_stiffness
    num_class_stiffness = 3

    use_roughness = args.use_roughness
    num_class_roughness = 3

    p_threshold = args.p_threshold

    # loss weightage
    w_r = args.w_r
    w_c = 1
    w_p = args.w_p

    np.random.seed(args.rand_seed)
    torch.manual_seed(args.rand_seed)

    # Load data
    # data_dir = os.path.join(args_data_dir, "compiled_data/")
    logDir = 'models_and_stats/'
    if_plot = False

    # RNN block
    block = "LSTM" # LSTM, GRU, phased_LSTM

    model_name = 'p_in_lat_stiffness{}_roughness{}_p_thre_{}_rs_{}_B_block_{}_data_{}_wrI_{}_wC_{}_wp_{}_hidden_{}_latent_{}_k_{}'.format(str(use_stiffness), str(use_roughness), str(p_threshold), str(args.rand_seed), block, dataset, w_r, w_c, w_p, str(hidden_size), str(latent_length), str(kfold_number))

    if torch.cuda.is_available() and cuda:
        device = torch.device("cuda:{}".format(args.cuda))
    else:
        device = torch.device('cpu')

    if args.reduction != 1:
        print("load {} kfold number, reduce data to {} folds, put to device: {}".format(args.kfold, args.reduction, device))
    else:
        print("load {} kfold number, train with full data, put to device: {}".format(args.kfold, device))

    prefix = ""
    # dataset_dir = os.path.join(args_data_dir, dataset+"/") # TODO
    # train_set, val_set, test_set = get_TrainValTestDataset(dataset_dir, k=kfold_number, prefix=prefix, seq_len=sequence_length)
    # train_loader, val_loader, test_loader = get_TrainValTestLoader(dataset_dir, k=kfold_number, batch_size=batch_size,shuffle=shuffle, prefix=prefix,seq_len=sequence_length)
    # X_train, X_val, X_test, Y_train, Y_val, Y_test = get_TrainValTestData(dataset_dir, k=kfold_number, prefix=prefix,seq_len=sequence_length)
    c30_x = np.load('data/c30/bio_all.npy')
    c30_y = np.load('data/c30/label_all.npy')
    c30_x = torch.FloatTensor(c30_x)
    c10_x, c10_y, stiff_y, rough_y = c30_process_data(c30_x, c30_y)
    print(c10_y[10:], stiff_y[10:])

    # Initialize models
    model = VRAEC(num_class=num_class_texture,
                block=block,
                sequence_length=sequence_length, # TODO
                number_of_features = number_of_features,
                hidden_size = hidden_size, 
                hidden_layer_depth = hidden_layer_depth,
                latent_length = latent_length,
                batch_size = batch_size,
                learning_rate = learning_rate,
                n_epochs = n_epochs,
                dropout_rate = dropout_rate,
                cuda = cuda,
                model_name=model_name,
                header=header,
                device = device)
    model.to(device)

    # Initialize training settings
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    cl_loss_fn = nn.NLLLoss()
    recon_loss_fn = nn.MSELoss()

    # model.load_state_dict(torch.load('models_and_stats/model_phased_LSTM_B30.pt', map_location='cpu'))
    saved_dicts = torch.load('models_and_stats/'+model_name+str(99)+'.pt', map_location='cpu')
    model.load_state_dict(saved_dicts['model_state_dict'])
    optimizer.load_state_dict(saved_dicts['optimizer_state_dict'])

    training_start=datetime.now()
    # create empty lists to fill stats later
    epoch_train_loss = []
    epoch_train_acc = []
    epoch_val_loss = []
    epoch_val_acc = []
    epoch_train_stiffness_loss = []
    epoch_train_stiffness_acc = []
    epoch_val_stiffness_loss = []
    epoch_val_stiffness_acc = []
    epoch_train_roughness_loss = []
    epoch_train_roughness_acc = []
    epoch_val_roughness_loss = []
    epoch_val_roughness_acc = []
    max_val_acc = 0
    max_val_epoch = 0
    if block == "phased_LSTM":
        time = torch.Tensor(range(sequence_length))
        times = time.repeat(batch_size, 1)

    # model.is_fitted = True
    model.eval()

    # TEST at the best model
    correct_texture = 0
    test_num = 0
    correct_stiffness = 0
    correct_roughness = 0
    # saved_dicts = torch.load('models_and_stats/'+model_name+str(max_val_epoch)+'.pt', map_location='cpu')
    # model.load_state_dict(saved_dicts['model_state_dict'])

    x_decoded, latent, output_texture = model(c10_x)
    correct_stiffness += torch.sum((torch.abs(latent[:, 0] - stiff_y) < p_threshold))
    correct_roughness += torch.sum((torch.abs(latent[:, 1] - rough_y) < p_threshold))
    pred_stiffness = np.zeros_like(stiff_y)
    _stiff_y = stiff_y
    for (label_ct, label_dc) in [(-1,-1), (1,1), (0, 0)]:
        pred_stiffness[torch.abs(latent[:, 0] - label_ct) < p_threshold] = label_dc
        _stiff_y = np.where(_stiff_y==label_ct, label_dc, _stiff_y)
    pred_roughness = np.zeros_like(rough_y)
    _rough_y = rough_y
    for (label_ct, label_dc) in [(-1,-1), (1,1), (0, 0)]:
        pred_roughness[torch.abs(latent[:, 1] - label_ct) < p_threshold] = label_dc
        _rough_y = np.where(_rough_y==label_ct, label_dc, _rough_y)
    test_num = c10_x.size(0)
    print('test_num', test_num)
    wrong_stiff_index = [(index, c10_y[index].item(), stiff_y[index].item(), pred_stiffness[index].item()) for index,value in enumerate(pred_stiffness) if stiff_y[index] != value]
    print('wrong_stiff_index, (index, y[index].item(), stiff_y[index].item(), pred_stiffness[index].item())\n', wrong_stiff_index)
    wrong_rough_index = [(index, c10_y[index].item(), rough_y[index].item(), pred_roughness[index].item()) for index,value in enumerate(pred_roughness) if rough_y[index] != value]
    print('wrong_rough_index, (index, y[index].item(), rough_y[index].item(), pred_roughness[index].item())\n', wrong_rough_index)

    confusion_matrix_show(stiff_y, pred_stiffness)
    confusion_matrix_show(rough_y, pred_roughness)
    
    test_stiffness_acc2 = correct_stiffness.item() / test_num
    test_roughness_acc2 = correct_roughness.item() / test_num
    # print('at the best model Test accuracy for', str(kfold_number), ' fold : ', test_acc2, test_stiffness_acc2, test_roughness_acc2)
    print('Transfer Test accuracy for', str(kfold_number), ' fold : ', test_stiffness_acc2, test_roughness_acc2)

    return test_stiffness_acc2, test_roughness_acc2
```

```python
# dummy class to replace argparser, if running jupyter notebook
class Args:
    rep = 0
    data_dir = 'data'
    kfold = 0
    cuda = '0'
    reduction = 1
    r_on = 0.1
    p_max = 200
    w_r = 0.01
    w_p = 1.0
    h_s = 90
    dataset = 'c20'
    stiffness_lat = 10
    roughness_lat = 10

args=Args()
args.latent_length = 40
args.rand_seed = 1
result = []
args.w_r = 0.01
args.p_threshold = 0.25
args.sequence_length = 400

args.dataset = 'c20'
args.use_stiffness = 1
args.use_roughness = 1
for w_p in [2,5,10]:
    args.w_p = w_p
    for k in range(5):
        args.kfold = k
        print(args)
        test_stiffness_acc1, test_roughness_acc1 = train(args)
        result.append((test_stiffness_acc1, test_roughness_acc1))
print(result)                       
```
